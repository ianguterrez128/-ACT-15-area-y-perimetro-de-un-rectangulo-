* Desarrollar un programa interactivo en Python, orientado a estudiantes de nivel preparatoria, que permita calcular propiedades geométricas básicas de un rectángulo mediante un menú en consola.

* El sistema ofrecerá tres opciones: calcular el área, calcular el perímetro o salir del programa. Al seleccionar cualquier operación, el programa solicitará de forma independiente los valores de la base y la altura, aplicará la fórmula matemática correspondiente y mostrará el resultado con dos decimales.
* El menú se mantendrá activo mediante un ciclo repetitivo hasta que el usuario decida finalizar, utilizando funciones modulares (arearectangulo(), perimetrorectangulo() y menu()) y estructuras de control while, if, elif y else.
* El código se organizará en un único archivo, sin emplear programación orientada a objetos ni validación de datos, y priorizará una interfaz clara y atractiva mediante emojis y formato de texto, solicitando las dimensiones únicamente al momento de ejecutar cada función.

```python
# ======================
# PROGRAMACION DE LOS CALCULOS GEOMETRICOS
# elaborado por ian gutierrez
# formulas area del circulo, perimetro del circulo, volumen de la esfera


import math


def arearectangulo():
    """Calcula el área de un rectángulo solicitando base y altura dentro de la función"""
    print("\n📐 ─── CÁLCULO DEL ÁREA DEL RECTÁNGULO ───")
    base = float(input("🔹 Ingresa la base del rectángulo: "))
    altura = float(input("🔹 Ingresa la altura del rectángulo: "))
    area = base * altura
    print(f"✅ Resultado: El área es {area:.2f} unidades²\n")

def perimetrorectangulo():
    """Calcula el perímetro de un rectángulo solicitando base y altura dentro de la función"""
    print("\n📏 ─── CÁLCULO DEL PERÍMETRO DEL RECTÁNGULO ───")
    base = float(input("🔹 Ingresa la base del rectángulo: "))
    altura = float(input("🔹 Ingresa la altura del rectángulo: "))
    perimetro = 2 * (base + altura)
    print(f"✅ Resultado: El perímetro es {perimetro:.2f} unidades\n")

def menu():
    """Función principal que muestra el menú y controla la ejecución del programa"""
    while True:
        print("\n" + "✨" * 20)
        print("   🎓 MENÚ GEOMÉTRICO 🎓")
        print("✨" * 20)
        print("1️⃣  Calcular área rectángulo")
        print("2️⃣  Calcular perímetro rectángulo")
        print("3️⃣  Salir")
        print("✨" * 20)

        opcion = int(input("👉 Selecciona una opción (1-3): "))

        if opcion == 1:
            arearectangulo()
        elif opcion == 2:
            perimetrorectangulo()
        elif opcion == 3:
            print("\n👋 ¡Gracias por usar el programa! Hasta pronto. 🌟")
            break
        else:
            print("⚠️ Opción no válida. Por favor, selecciona un numero del 1 al 3")
            print("programa realizado por ian gutierrez")

# Punto de entrada del programa
if __name__ == "__main__":
    menu()
```
